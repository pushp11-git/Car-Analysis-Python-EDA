Q) Create a histogram of the 'MPG_City' values.

```python
import matplotlib.pyplot as plt
plt.hist(car['MPG_City'], bins=15, edgecolor='black')
plt.title('Histogram of MPG_City')
plt.xlabel('MPG_City')
plt.ylabel('Frequency')
plt.show()
```

Plot a scatter plot of 'EngineSize' vs 'Horsepower' and color the points by 'TYpe'.

```python
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for handling potential NaN values

origin_colors = {'Sports': 'blue', 'SUV': 'green', 'Sedan': 'red','Truck' : 'yellow', 'Wagon': 'cyan', 'Hybrid':'orange' }
colors = car['Type'].map(origin_colors).fillna('gray')  # Map colors and fill NaN with gray
plt.scatter(car['EngineSize'], car['Horsepower'], c=colors.tolist())  # Convert colors to a list
plt.title('EngineSize vs Horsepower')
plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.show()
```

Create a line plot showing the average 'MSRP' over different 'Type'.

```python
avg_msrp_by_cylinders = car.groupby('Type')['MSRP'].mean()
plt.plot(avg_msrp_by_cylinders.index, avg_msrp_by_cylinders.values)
plt.title('Average MSRP by Type')
plt.xlabel('Types')
plt.ylabel('Average MSRP')
plt.show()
```

Generate a box plot to visualize the distribution of 'Weight' across different 'Type' categories.

```python
car.boxplot(column='Invoice', by='Type', grid=False)
plt.title('Box Plot of Weight by Type')
plt.suptitle('')
plt.xlabel('Type')
plt.ylabel('Invoice')
plt.show()
```

Plot a pie chart to show the proportion of car types in the dataset.
